```python
# Define constants for gender classification
MALE = 'male'
FEMALE = 'female'
UNKNOWN = 'unknown'
NAME_DATA = 'names_data.txt'

# Lists of gender-specific words
MALE_WORDS = set(['guy', 'spokesman', 'chairman', "men's", 'him', "he's", 'his',
              'boy', 'boyfriend', 'boyfriends', 'boys', 'brother', 'brothers', 'dad',
              'dads', 'dude', 'father', 'fathers', 'fiance', 'gentleman', 'gentlemen',
              'god', 'grandfather', 'grandpa', 'grandson', 'groom', 'he'
])

FEMALE_WORDS = set(['heroine', 'spokeswoman', 'chairwoman', "women's", 'actress'
])

# Function to determine gender based on words
def genderize(words):
    mwlen = len(MALE_WORDS)
    fwlen = len(FEMALE_WORDS)
    if mwlen > 0 and fwlen > 0:
        m = sum(1 for word in words if word in MALE_WORDS) / mwlen
        f = sum(1 for word in words if word in FEMALE_WORDS) / fwlen
        return (MALE if m > f else FEMALE) if m != f else UNKNOWN

# Parse text and analyze gender distribution
def parse_gender(text):
    sentences = [
        [word.lower() for word in nltk.word_tokenize(sentence)]
        for sentence in nltk.sent_tokenize(text)
    ]
    sents, words = count_gender(sentences)
    total = sum(words.values())
    for gender, count in words.items():
        pcent = (count / total) * 100
        nsents = words[gender]
        print ("{:0.3f}% {} ({} sentences)".format(pcent, gender, nsents))

# Test our gender analysis
parse_gender("What is she doing?") # Should identify as female
```

```python
# Identifying unusual words
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words()) # set 2: the nltk dictionary
    unusual = text_vocab - english_vocab # compare the difference b/w 2 sets
    return sorted(unusual)
    print(unusual_words(text1))

unusual_words(text1)


# Calcualte what percentage of the text consists of content words
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
    print(content_fraction(text1))

content_fraction(text1) # Percentage of content words in Emma
```
